```python
import pandas as pd
```

```python
ksi=pd.read_csv("Data/ksi_cleaned.csv")
ksi.head(10)
```

```python
# Remove non-fatal injuries of fatal accidents from the dataframe
new_ksi = ksi.loc[((ksi.ACCLASS == "Fatal") & (ksi.INJURY == "Fatal")) | (ksi.ACCLASS == "Non-Fatal Injury")]
new_ksi.head(10)
```

```python
# Keep one record for each fatal or non-fatal accident
ksi_by_accident = new_ksi.drop_duplicates('ACCNUM')
ksi_by_accident.head(10)
```

```python
ksi_by_accident.to_csv("Data/ksi_by_accident.csv",index=False)
```
